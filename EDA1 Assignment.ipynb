```python
##EDA on Cardiographic Dataset
##import libraries and load data
import pandas as pd
import numpy as np
import seaborn as sns
df=pd.read_csv("cardiotocographic.csv")
```

```python
df
```

```python
##check data info and clean it
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.apply(pd.to_numeric,errors='coerce')
df
```

```python
df=df.fillna(df.mean())
```

```python
df
```

```python
##Detect and handle outliers using IQR
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
df=df[~((df<(Q1-1.5*IQR)) | (df>(Q3+1.5*IQR))).any(axis=1)]
df.describe()
```

```python
##show summary statistics
df.describe()
```

```python
##visualize the data
##histogram
import matplotlib.pyplot as plt
df.hist(figsize=(12,8),bins=15)
plt.tight_layout()
plt.show()
```

```python
##Boxplot
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title("Boxplot of all features")
plt.show()
```

```python
##correlation Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
##Violin plot example
plt.figure(figsize=(10,6))
sns.violinplot(data=df[['LB','AC','FM']])
plt.title("Violin plots of LB,AC and FM")
plt.show()
```

```python
##pattern recognition and insights
##Identify top correlations
correlation_matrix=df.corr()
strong_corr=correlation_matrix[(correlation_matrix>0.5)&(correlation_matrix>0.1)]
print(strong_corr.dropna(how='all').dropna(axis=1,how='all'))
sns.scatterplot(x='LB',y='AC',data=df)
plt.title("LB VS AC")
plt.show()
```

```python
##Conclusion and Recommendations
Key Findings:

The dataset is clean with no major missing values after preprocessing.

LB, AC, and UC are key indicators of fetal well-being.

Variability indicators like ASTV, ALTV help assess fetal health variability.

Outliers were present in a few deceleration measures and were removed using IQR.

Recommendations:
Use LB, AC, and ASTV as input features for potential classification models.

Consider further medical interpretation before making decisions.

Explore clustering or classification (e.g., healthy vs. distressed) in future analysis.
```
