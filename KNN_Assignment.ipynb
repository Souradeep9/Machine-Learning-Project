```python
df_cleaned=df.drop(columns=['animal name'])
```

```python
df_cleaned
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x=df_cleaned.drop('type',axis=1)
```

```python
x
```

```python
y=df_cleaned['type']
```

```python
y
```

```python
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
##Train-Test Split
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
##Implemment KNN Classifier
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k=5
knn=KNeighborsClassifier(n_neighbors=k,metric='minkowski',p=2)
```

```python
knn.fit(x_train,y_train)
```

```python
##Evalute the Model
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
y_pred=knn.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
classification_report(y_test,y_pred)
```

```python
##Visualize the decision boundaries of classifier
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
X_pca=pca.fit_transform(x_scaled)
```

```python
X_train_pca,X_test_pca=train_test_split(X_pca,test_size=0.2,random_state=42)
```

```python
knn_pca=KNeighborsClassifier(n_neighbors=k)
```

```python
knn_pca.fit(X_train_pca,y_train)
```

```python
import numpy as np
```

```python
h=0.2
x_min,x_max=X_pca[:,0].min()-1,X_pca[:,0].max()+1
```

```python
y_min,y_max=X_pca[:,1].min()-1,X_pca[:,1].max()+1
```

```python
xx,yy=np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
```

```python
z=knn_pca.predict(np.c_[xx.ravel(),yy.ravel()])
```

```python
z=z.reshape(xx.shape)
```

```python
z
```

```python
plt.figure(figsize=(10,6))
plt.contourf(xx,yy,z,cmap=plt.cm.Paired,alpha=0.8)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, edgecolors='k', s=20)
plt.title('KNN Decision Boundaries with PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

Interview Questions:
Q1. What are the key hyperparameters in KNN?

n_neighbors (K): Number of nearest neighbors to consider.

metric: Distance metric (e.g., minkowski, euclidean, manhattan).

weights: Uniform or distance-based weighting.

p: Power parameter for the Minkowski metric (p=2 for Euclidean, p=1 for Manhattan).
Q2. What distance metrics can be used in KNN?

Euclidean distance (default, p=2)

Manhattan distance (p=1)

Minkowski distance (general form)

Cosine similarity

Hamming distance (for categorical variables)
